## Что такое Leove-one-out? Привести плюсы и минусы:
Метод leove-one-out based on K-fold.
K-fold - это метод кросс-валидации, который позволяет выбирать гиперпараметры модели, которые наиболее эффективно влияют на её производительность и качество.
Как работает этот метод?
1. Формируется целое число K, меньшее числа семплов в датасете.
2. Датасет разбивается на K одинаковых частей (В последней части может быть меньше семплов). Эти части называются фолдами.
3. Далее происходит K-итераций, во время которых:
   а) Формируется $$x_i$$ фолд - тестовый набор данных
   б) Формируется и объединяется $$x_i, ... x_{i-1}, x_{i+1}, ... x_k$$ фолдов - тренировочный набор данных.
   в) Модель учится на $$k-1$$ фолде и тестируется на оставшемся.
4. Финальный скор модели получется:
   - Усреднением $K$ получившихся тестовых результатов
     или
   - Измерением на отложенном тестовом множестве, которое не участвовало в кросс-валидации
После применения метода k-fold для одной модели результатом будет k-инстансов этой модели, обученной на разных подмножествах трейна.
Че делать?
1. делать предсказания на основе усреднения предсказаний каждого k-инстанса
2. Выбирать тот инстанс, который набрал наибольший скор на тестовых данных/тестовом множестве
3. Заново обучить модель уже на всех k-фолдах и делать предсказания уже этой моделью
Плюсы - надежность, т.к. обучения и тест модели присходят на разных подмножествах исходного датаеста
Минусы - Вычислительно затратно. Следует применять только при большом количестве вычислительных ресурсов, либо при малом количестве данных.

Честным примером K-fold является levave-one-out.
Каждый фолд состоит ровно из одного семпла(k = ```len(X_train)```).
Его идеально использовать при малом количестве данных.
Однако из-за построения примерров по одному является ресурсозатратным, так как на каждой итерации модели валидация идет по одному семплу.

## Grid Search, Randomized Grid Search, Bayesian optimization
1. Grid Search
Пусть для модели доступно $m$ гиперпараметров
В диапазоне $[min_{i}, max_{i}]$ выбирается $k_{i}, i = 1, m$ различных допустимых значений гиперпараметра. Тогда количество гиперпараметров оценивается как число различных комбинаций элементов из m множеств с мощностями $k_1, k_2,...k_,$:
рисуночек сеточки
2. Randomized Grid Search
Вместо перебора по сетке можно случайным образом выбирать значения гиперпараметров из допустимого множества значений.
рисуночек и формула
Случайный поиск позволяет быстрее - за меньшее число итераций, - найти оптимальную область значений гиперпараметра.
3. Bayesian optimization
Пусть гипермараметры $phi^m$ влияют на значения оптимизируемого функционала q( ну как, пусть, они реально влияют, бля буду)
Значит существует условие $p_{true} = (q | phi)$, которое неизвестно.
Тогда при каждой попытке обучения в качестве результата получается узнать для $phi_i$ значение $q_i$.
На основе этой информации можно построить вероятность апостериорного распределения $p_{approx} = (q | phi)$
(Апостериорное распределения задается параметрически и оптимизируют его параметры)
И уже на основе этого распределения можно будет найти точки минимума и максимума.
Как итоге - каждый шаг подбора гиперпараметров дорожно, но и количество итераций уменьшается в разы

### a) Split data into 2 parts randomly with parameter test_size (ratio from 0 to 1), return training and test samples.
### b) Randomly split data into 3 parts with parameters validation_size and test_size, return train, validation and test samples.

````python
def validate_variables(func):
    @wraps(func)
    def wrapper(*args, **kwargs):

        # проверка для date:
        date_split = kwargs.pop("date_split", None)
        test_date = kwargs.pop("test_date", None)
        val_date = kwargs.pop("val_date", None)
        if (
            (date_split is not None)
            or (test_date is not None and val_date is not None)
            or (test_date is not None)
        ):
            if date_split is not None and test_date is not None:
                raise ValueError(
                    "Use 'date_split' or 'test_data' - using both is denied"
                )
            final_test_point = date_split or test_date
            if final_test_point is None:
                raise ValueError("Required point for split test")
            try:
                kwargs["test_date"] = pd.to_datetime(final_test_point)
                if val_date is not None:
                    kwargs["val_date"] = pd.to_datetime(val_date)
            except (ValueError, TypeError):
                raise TypeError(
                    f"Invalid date format. Failed to parse: {final_test_point or val_date}"
                )

        arrays = args
        if not arrays:
            raise ValueError("You should add at least one array to the func")
        n_samples = len(arrays[0])

        validated_arr = []

        for arr in arrays:
            if not hasattr(arr, "__len__"):
                raise ValueError(
                    f"Expected: array-like data structure \n Recieved: {type(arr)}- неведомая хрень"
                )
            if len(arr) != n_samples:
                raise ValueError(f"Diffrent len of arrays: [{n_samples}, {len(arr)}]")

            # проверка для date
            if kwargs.get("test_date") is not None:
                if (
                    kwargs.get(
                        "date_column",
                    )
                    is None
                ):
                    raise ValueError("expected date_column")
                if not isinstance(kwargs["date_column"], str):
                    raise TypeError(
                        f"date_column expectes str, recieved {type(kwargs['date_column'])}"
                    )

                if kwargs["date_column"] not in arr.columns:
                    raise KeyError(f"Data required {kwargs['date_column']} column")
                arr = arr.sort_values(by=kwargs["date_column"])
            # че то надо тут придумать, а то как-то не красивовое
            if isinstance(arr, (pd.Series, pd.DataFrame)):
                validated_arr.append(arr)
            else:
                validated_arr.append(arr)
        for param in ["test_size", "val_size"]:
            if param in kwargs:
                val = kwargs[param]
                if not isinstance(val, (float)):
                    raise TypeError(
                        f"Params {param} should be a float, been getted: {type(val)}"
                    )
                if not (0.0 < val < 1.0):
                    raise ValueError(
                        f"Parametr {param} should be in range (0, 1), been getted: {val}"
                    )
        return func(*validated_arr, **kwargs)

    return wrapper
````

````python
@validate_variables
def train_test_split_custom(
    *arrays,
    test_size=None,
    val_size=None,
    random_state=None,
    shuffle: bool = True,
    custom_indices: list = None
):
    n_arrays = len(arrays)
    if n_arrays == 0 or n_arrays > 2:
        raise ValueError("Incorrect count of input arrays")
    result = []
    n_samples = len(arrays[0])
    n_test = int(np.ceil(n_samples * test_size))
    if val_size is not None:
        n_val = int(np.ceil(n_samples * val_size))
        n_train = n_samples - n_test - n_val
        n_train, n_test, n_val = int(n_train), int(n_test), int(n_val)
    else:
        n_train = n_samples - n_test
        n_train, n_test = int(n_train), int(n_test)
    if n_train == 0:
        raise ValueError(
            "With n_samples={}, test_size={}, the "
            "resulting train set will be empty. Adjust any of the "
            "aforementioned parameters.".format(n_samples, test_size)
        )

    idx = np.arange(n_samples)

    if shuffle:
        np.random.seed(random_state) if random_state is not None else None
        np.random.shuffle(idx)

    train_random_idx = idx[:n_train]
    if val_size is not None:
        val_random_idx = idx[n_train : n_train + n_val]
        test_random_idx = idx[n_train + n_val :]
    else:
        test_random_idx = idx[n_train:]
    for i in range(n_arrays):
        result.append(arrays[i][train_random_idx])
        if val_size is not None:
            result.append(arrays[i][val_random_idx])
            result.append(arrays[i][test_random_idx])
        else:
            result.append(arrays[i][test_random_idx])
    return result
````

````python
@validate_variables
def data_split_custom(*arrays, date_column, test_date, val_date=None):
    results = []

    for arr in arrays:
        if val_date is None:
            results.append(arr.loc[arr[date_column] < test_date])
            results.append(arr.loc[arr[date_column] >= test_date])
        else:
            results.append(arr.loc[arr[date_column] < val_date])
            results.append(
                arr.loc[arr[date_column].between(val_date, test_date, inclusive="left")]
            )
            results.append(arr.loc[arr[date_column] >= test_date])

    return results


@validate_variables
def data_split_custom_without_date(*arrays, date_column, test_date, val_date=None):
    data = data_split_custom(
        *arrays, date_column=date_column, test_date=test_date, val_date=val_date
    )
    for arr in data:
        arr.drop(columns=[date_column], inplace=True)
    return data
````

````python
class BaseKFold_custom:
    def __init__(self, n_splits=5, random_state=None, shuffle: bool = False):
        self.n_splits = n_splits
        self.random_state = random_state
        self.shuffle = shuffle

    # @validate_variables
    def split(self, X, y=None, groups=None):
        n_samples = len(X)
        if self.n_splits > n_samples:
            raise ValueError(
                f"N_splits cannot be a greater then len(X): {[self.n_splits, n_samples]}"
            )

        indices = np.arange(n_samples)
        for test_index in self._iter_test_indices(X, y, groups):
            train_index = np.setdiff1d(indices, test_index, assume_unique=True)
            yield train_index, test_index

    def _iter_test_indices(self, X, y=None, groups=None):
        raise NotImplementedError("This method is not implemented")
````

````python
class KFold_custom(BaseKFold_custom):
    def _iter_test_indices(self, X, y=None, groups=None):
        n_samples = len(X)
        indices = np.arange(n_samples)
        if self.shuffle:
            np.random.default_rng(self.random_state).shuffle(indices)
        fold_sizes = np.full(self.n_splits, n_samples // self.n_splits, dtype=int)
        fold_sizes[: n_samples % self.n_splits] += 1
        current = 0
        for fold_size in fold_sizes:
            yield indices[current : current + fold_size]
            current += fold_size
````

### c) Stratified K-fold, where k and stratify_field are input parameters, returns list of train and test indices.

Концепция:
1. сгруппировать индексы таргета по меткам/классам/группам
2. внутри каждой группы нарезать на k частей
3. склеить k-тые между собой

````python
class StratifiedKFold_custom(BaseKFold_custom):
    def split(self, X, y, stratify_field=None):
        if stratify_field is None:
            raise ValueError("Stratify_field is required")
        return super().split(X=X, y=y, groups=stratify_field)

    def _iter_test_indices(self, X, y=None, groups=None):
        if groups is None:
            raise ValueError("Param groups cannot be a None")
        n_samples = len(X)
        if n_samples != len(groups):
            raise ValueError(
                "grops should be a same len as a X, recieved %d %d"
                % (n_samples, len(groups))
            )
        stratify_field = np.asarray(groups)
        unique_labels = np.unique(stratify_field)
        per_class_indices = [
            np.where(stratify_field == label)[0] for label in unique_labels
        ]

        if self.shuffle:
            rng = np.random.default_rng(self.random_state)
            for indices in per_class_indices:
                rng.shuffle(indices)

        per_class_splits = [
            np.array_split(indices, self.n_splits) for indices in per_class_indices
        ]

        for i in range(self.n_splits):
            # Из каждого класса берем i-тую часть и объединяем в один тест-сет
            test_indices = np.concatenate([splits[i] for splits in per_class_splits])
            yield test_indices
````

### d) Time series split, where k and date_field are input parameters, returns list of train and test indices.

````python
class TimeSeries_custom(BaseKFold_custom):
    def split(self, X, y=None, date_field=None):
        """Сортировка индексов по датам"""
        if date_field is None:
            raise ValueError("Для TimeSeriesSplit необходимо указать date_field")

        dates = pd.to_datetime(date_field)
        sort_indices = np.argsort(dates.values)

        return super().split(X, y, groups=sort_indices)

    def _iter_test_indices(self, X, y=None, groups=None):
        n_samples = len(X)
        step_size = n_samples // (self.n_splits + 1)

        indices = groups

        for i in range(1, self.n_splits + 1):
            # Конец тренировочной выборки (1/k, 2/k, ...)
            train_end = i * step_size

            test_start = train_end
            test_end = (i + 1) * step_size if i != self.n_splits else n_samples

            yield indices[test_start:test_end]
````
